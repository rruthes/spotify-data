## Importando as dependências.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
```

## Conferindo se há relação entre ano de lançamento da música e número de streams.

```python
df = pd.read_csv('../data/processed/top50_streams.csv')

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='released_year', palette='viridis')

plt.title('Anos de Lançamento das Músicas Mais Tocadas')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Quantidade de Músicas')

plt.show()
```

## Conferindo se há relação entre mês de lançamento da música e número de streams.

```python
df = pd.read_csv('../data/processed/top50_streams.csv')

df['released_month'] = df['released_month'].astype('category')

streams_mes = df.groupby('released_month')['streams'].sum()

plt.figure(figsize=(10, 8))
plt.pie(streams_mes, labels=streams_mes.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', 12))

plt.title('Distribuição de Streams por Mês de Lançamento')

plt.show()
```
