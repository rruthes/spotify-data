Importando as dependÃªncias.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
```

Transformando o conjunto geral de dados em um data set e limpando os dados.

```python
df = pd.read_csv('../data/raw/Spotify Most Streamed Songs.csv')
df.dropna()
df = df.drop(columns=['cover_url', 'key', 'in_shazam_charts', 'in_deezer_charts', 'in_deezer_playlists', 'mode'])
df.head(50)
```

Listando as mÃºsicas com o maior nÃºmero de streams, em ordem decrescente.

```python
df = df[df['track_name'] != 'Love Grows (Where My Rosemary Goes)']
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
df_sorted = df.sort_values(by='streams', ascending=False)
p_streams = df_sorted.head(50)
p_streams.to_csv('../data/processed/top50_streams.csv', index=False)
p_streams.head(10)
```

Listando as mÃºsicas que mais estÃ£o em playlists no Spotify.

```python
df['in_spotify_playlists'] = pd.to_numeric(df['in_spotify_playlists'], errors='coerce')
df_sorted = df.sort_values(by='in_spotify_playlists', ascending=False)
p_playlists = df_sorted.head(50)
p_playlists.to_csv('../data/processed/top50_playlists.csv', index=False)
p_playlists.head(10)    
```

Listando as que mais estÃ£o em playlists do Apple Music.

```python
df['in_apple_playlists'] = pd.to_numeric(df['in_apple_playlists'], errors='coerce')
df_sorted = df.sort_values(by='in_apple_playlists', ascending=False)
p_apple_playlists = df_sorted.head(50)
p_apple_playlists.to_csv('../data/processed/top50_apple_playlists.csv', index=False)
p_apple_playlists.head(10)
```
