Importando as dependências.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
```

Transformando o conjunto geral de dados em um data set.

```python
df = pd.read_csv('../data/raw/Spotify Most Streamed Songs.csv')
df.head()
```

Listando as músicas com o maior número de streams, em ordem decrescente.

```python
df = df[df['track_name'] != 'Love Grows (Where My Rosemary Goes)']
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
df_sorted = df.sort_values(by='streams', ascending=False)
p_streams = df_sorted.head(10)
p_streams.to_csv('../data/processed/top10_streams.csv', index=False)
df_sorted.head(10)
```

Listando as músicas que mais estão em playlists no Spotify.

```python
df['in_spotify_playlists'] = pd.to_numeric(df['in_spotify_playlists'], errors='coerce')
df_sorted = df.sort_values(by='in_spotify_playlists', ascending=False)
p_playlists = df_sorted.head(10)
p_playlists.to_csv('../data/processed/top10_in_playlists.csv', index=False)
df_sorted.head(10)
```
